```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.options.display.max_rows = 10
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('D:\\Jupyter\\利用Python进行数据分析2\\yjx\\Python-\\chapter14\\data\\movielens\\users.dat',
                      sep='::',header=None, names=unames)
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('D:\\Jupyter\\利用Python进行数据分析2\\yjx\\Python-\\chapter14\\data\\movielens\\ratings.dat',
                        sep='::',header=None, names=rnames)
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('D:\\Jupyter\\利用Python进行数据分析2\\yjx\\Python-\\chapter14\\data\\movielens\\movies.dat',
                       sep='::',header=None, names=mnames)
users[:5]
```

```python
data = pd.merge(pd.merge(ratings,users),movies)
data
```

```python
mean_ratings = data.pivot_table('rating',index='title',columns='gender',aggfunc='mean')
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
top_female_ratings
```
